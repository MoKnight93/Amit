### Task 1
Suppose you have this email address “Amit_ml@gmail.edu”
- Input Validation: Check if the input string contains exactly one "@" symbol and
at least one "." after the "@" symbol. If it's not a valid email, return "Invalid
email".
- Extract Username: Extract and return the part of the email before the "@"
symbol.
- Extract Domain: Extract and return the domain (the part between "@" and the
last ".").
- Check for Domain Ending: Check if the email ends with ".com". If it does,
return "Commercial Domain". If it ends with ".edu", return "Educational Domain".
Otherwise, return "Other Domain

```python
e = "Amit_ml@gmail.edu"

l = e.find("@")

if e.count("@") == 1 and "." in e[l:]:
        print("Valid email")
else:
        print("Invalid email")
```

### Task 2

Encoded Message:
###!!@mocleW EPGTQ!!!6789

```python
s = "###!!@mocleW EPGTQ!!!6789"
s = s.replace("@", "e")
s1 = ""

for i in s:
    if i.isalpha() or i.isspace():
        s1 += i

l = s1.split()

l[0] = l[0][::-1]

print(l[0], l[1])
```

### Task 3
Encoded Message:
&&&**$gnirtS PLIO!!@1234

```python
s = "&&&**$gnirtS PLIO!!@1234"
s1 = ""

for i in s:
    if i.isalpha() or i.isspace():
        s1 += i

l = s1.split()

l[0] = l[0][::-1]

l[1] = l[1].replace("I", "E")
l[1] = l[1].replace("O", "U")

print(l[0], l[1])
```

### Task 4
Encoded Message:
##$$$@!yalpstcejorp EPUVT****9887

```python
s = "##$$$@!yalpstcejorp EPUVT****9887"
s1 = ""

for i in s:
    if i.isalpha() or i.isspace():
        s1 += i

l = s1.split()

l[0] = l[0][::-1]

l[1] = l[1].replace("E", "A")
l[1] = l[1].replace("U", "O")

print(l[0], l[1])
```
