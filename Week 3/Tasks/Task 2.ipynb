Task 2 – Thanos Project
Objective:
Create a Python function that performs a series of file and folder operations, inspired by the
concept of “Thanos” randomly removing half of the files.
Requirements
Write a function that does the following steps:
1. Open a folder
o Access an existing folder on your system.
2. Create files inside the folder
o Generate a specified number of files inside the folder.
3. Count the files
o Check and display the total number of files in the folder.
4. Delete half of the files randomly
o Randomly select and delete half of the files inside the folder.
5. Check the result
o Recount and display the remaining number of files after deletion.
Notes
• File deletion must be random.
• The function should print clear messages showing each step and its result.
• Use Python standard libraries such as os and random.
Expected Outcome
After running the function:
• Files are created successfully.
• Half of them are randomly removed.
• The final number of files is displayed correctly.

```python
import os
import random

def thanos_project():
    """
    Thanos project: Delete random half of files in a folder
    """
    
    # Step 1: Open the whole folder
    folder_name = "thanos_folder"
    
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    print("Step 1: Folder opened/created")
    
    # Step 2: Create number of files inside it
    num_files = 10
    
    for i in range(1, num_files + 1):
        with open(f"{folder_name}/file_{i}.txt", 'w') as f:
            f.write(f"File {i}")
    
    print(f"Step 2: Created {num_files} files")
    
    # Step 3: Check the len of files in that folder
    files = os.listdir(folder_name)
    print(f"Step 3: Total files = {len(files)}")
    
    # Step 4: Delete random half of this folder
    half = len(files) // 2
    files_to_delete = random.sample(files, half)
    
    for file in files_to_delete:
        os.remove(f"{folder_name}/{file}")
    
    print(f"Step 4: Deleted {half} files")
    
    # Step 5: Check the result
    remaining_files = os.listdir(folder_name)
    print(f"Step 5: Remaining files = {len(remaining_files)}")


# Run the function
thanos_project()
```
