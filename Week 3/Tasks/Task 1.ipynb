Task 1 – Simple Calculator Program
Objective:
Develop a Python program that acts as a simple calculator. The program should allow the user to
choose an arithmetic operation and apply it to two numbers entered by the user.
Description
The program is menu-driven and performs basic arithmetic operations using separate functions for
each operation. The user selects an operation, enters two numbers, and the program displays the
calculated result.
Requirements
1. Define arithmetic functions
Create separate functions for:
o Addition
o Subtraction
o Division
o Multiplication
2. Display a menu
o Show a menu that allows the user to choose one of the available operations.
3. User input
o Ask the user to enter:
▪ The desired operation
▪ Two numeric values
4. Perform the calculation
o Call the appropriate function based on the user’s choice.
o Handle invalid operation selections.
5. Display the result
o Print the result of the selected operation clearly.
6. Main program execution
o Use a main() function.
o Ensure the program runs only when executed directly.
Notes
• The program should use functions to keep the code organized.
• User input must be converted to numerical values before calculations.
• The output should clearly show the result.

```python
def sum(num_1, num_2):
    return num_1 + num_2

def sub(num_1, num_2):
    return num_1 - num_2

def div(num_1, num_2):
    return num_1 / num_2

def multiply(num_1, num_2):
    return num_1 * num_2

def main():
    while True:
        print("Python Calculator")
        print("-" * 20)
        print("Choose an operation: \n")
        print("1 - Addition (+)")
        print("2 - Subtraction (-)")
        print("3 - Division (÷)")
        print("4 - Multiplication (×)")
        print("5 - Exit")
        print("-" * 20)
        
        choice = input("Enter your choice (1-5): ")

        try:
            if choice == "5":
                print("Thank you for using the calculator!")
                break
        except KeyboardInterrupt:
            print("\nOperation cancelled by user. Exiting...\n")
            break
        except Exception as e:
            print(f"Unexpected error: {e}")
            break

        if choice not in ["1", "2", "3", "4"]:
            print("\nInvalid choice! Please select 1-4, or 5 to exit\n")
            continue

        try:
            num_1 = float(input("Enter first number: "))
            num_2 = float(input("Enter second number: "))
        except ValueError:
            print("\nError: Please enter valid numbers!\n")
            return
        except Exception as e:
            print(f"Unexpected error: {e}")
            return
        
        try:
            if choice == "3" and num_2 == 0:
                print("\nError: Division by zero is not allowed!\n")
                continue
        except ZeroDivisionError:
            print("Division Error")
        except Exception as e:
            print(f"Unexpected error during calculation: {e}")

        if choice == "1":
            result = sum(num_1, num_2)
            operation = "+"
        elif choice == "2":
            result = sub(num_1, num_2)
            operation = "-"
        elif choice == "3":
            result = div(num_1, num_2)
            operation = "÷"
        elif choice == "4":
            result = multiply(num_1, num_2)
            operation = "×"

        print(f"\nResult: {num_1} {operation} {num_2} = {result}\n")

if __name__ == "__main__":
    main()
            
```
